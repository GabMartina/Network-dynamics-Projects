**Exercise 1: Continuous-time random walk**

```python
G = nx.DiGraph()
G.add_edges_from([('o','a'),('o','b'),('a','b'),('a','c'),('b','o'),('b','c'),('c','b'),('c','d'),('d','c'),('d','a')])
pos = {'o':[0,1], 'a':[1,0], 'd':[2,0], 'b':[1,2], 'c':[2,2]}
nx.draw(G, pos, with_labels=True)
```
